```python
import matplotlib.pyplot as plt
```

Key components:
1. plotting area
2. legend
3. axis labels
4. ticks
5. title

```python
x = [1,2,3,4,5]
y = [2,3,5,7,11]
```

```python
plt.figure(figsize = (8,6))

plt.plot(x, y, label = "Prime numbers", marker = "*", markersize = 10, color="green", linestyle = "--", linewidth = 1)

plt.title("simple line plot")
plt.xlabel("x - axis")
plt.ylabel("y - axis")

plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
x = np.array([0,1,2,3,4,5])
```

```python
y = x ** 2
```

```python
y
```

```python
plt.plot(x, y, marker = "o", markersize = 5, linestyle = "--", linewidth = 4, color = "red", label = "data points")

plt.title("x vs y")
plt.xlabel("x values")
plt.ylabel("y values")

plt.grid(True)

plt.xticks(np.arange(0,6, step = 2))
plt.yticks(np.arange(0, 31, step=10))

plt.legend()

plt.show()
```

```python
purchase["purch_date"] = pd.to_datetime(purchase["purch_date"])
```

```python
purchase.dtypes
```

```python
purchase["month"] = purchase["purch_date"].dt.month
```

```python
purchase
```

```python
spend_month = purchase.groupby("month")["amount"].sum().reset_index()
```

```python
plt.figure(figsize=(10,8))

plt.plot(spend_month["month"], spend_month["amount"], marker="*", markersize = 5, linestyle = "--", linewidth = 2, color = "green")

plt.title("customer spend pattern by month")
plt.xlabel("months")
plt.ylabel("amount spent")

plt.xticks(np.arange(1,13, step = 1))

plt.grid(True)

plt.show()
```

## Seaborn

```python
# !pip install seaborn
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.barplot(x = "day", y = "tip", data = tips, estimator=sum)
plt.title("total tips by day")
plt.xlabel("day")
plt.ylabel("total tips")
plt.show()
```

```python
sns.scatterplot(x="total_bill", y = "tip", data = tips, hue = "time", style = "time")

plt.title("total bill vs tips")
plt.xlabel("total bill")
plt.ylabel("tips")
plt.show()
```

```python
sns.boxplot(x = "day", y = "total_bill", data = tips)
plt.title("box plots of total bill by days")
plt.xlabel("days")
plt.ylabel("total bill")
plt.show()
```
